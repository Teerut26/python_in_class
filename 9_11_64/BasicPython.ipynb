# **ตัวแปรและประเภทข้อมูล**

```python
a, b = 1, 2
x = y = z = 10
print("a = " , a)
print("b = " , b)
print("x = " , x)
print("y = " , y)
print("z = " , z)
```

```python
# Integer
a = 7
b = 3
c = a + b
d = a / b

print ('a = %d' % a)
print ('b = %d' % b)
print ('c = %d' % c)
print ('d = ', d)
```

```python
# Floating point number
speed = 34.12
pi = 22 / 7
height = 2.31E5
length = 1.3E-3

print ('speed = %f' % speed)
print ('pi = %f' % pi)
print ('height = %f' % height)
print ('length = %f' % length)
print (pi)
```

```python
sentent1 = "What's your name?"
sentent2 = 'I\'m Mateo.'
sentent3 = "He said \"I would learn Python first\"."
sentent4 = 'His teach replied "Oh well!"'
print (sentent1)
print (sentent2)
print (sentent3)
print (sentent4)
```

```python
site = 'Computer' + 'Visut'
tutorial = 'Python' ' Language'
print(site)
print(tutorial)
```

```python
a = 8
b = 13.4
c = "Python"
d = [1, 2, 3, 4]

print('Type of a = ', type(a))

print('Type of b = ', type(b))

print('Type of c = ', type(c))

print('Type of d = ', type(d))
```

# **โจทย์ปัญหา**

จงเขียนโปรแกรมเพื่อคำนวนค่าในสมการดังต่อไปนี้ โดยให้ผู้ใช้เป็นคนกรอกข้อมูลค่าในตัวแปลเอง (ควรศึกษาหาวิธีการใช้เครื่องหมายรากที่สอง)111.JPG

```python
import math
x1 = int(input("x1 : "))
x2 = int(input("x2 : "))
y1 = int(input("y1 : "))
y2 = int(input("y2 : "))

pos1 = ((x1 - x2) ** 2 )
pos2 = ((y1 - y2) ** 2 )
distance = math.sqrt(pos1+pos2)
print(distance)
```
