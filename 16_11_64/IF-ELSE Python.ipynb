## **if**

Python_if_statement.webp

```python
num = 3
if num > 0:
    print(num, "is a positive number.")
print("This is always printed.")

num = -1
if num > 0:
    print(num, "is a positive number.")
print("This is also always printed.")
```

# **if...else**

Python_if_else_statement.webp

```python
num = 3

if num >= 0:
    print("Positive or Zero")
else:
    print("Negative number")
```

# **if...elif...else**

```python
num = 3.4

if num > 0:
    print("Positive number")
elif num == 0:
    print("Zero")
else:
    print("Negative number")
```

# **โจทย์**

คุณเป็นวัยรุ่นหรือเปล่า โดยมีเงื่อนไขคำถามเพื่อใช้ในการตัดสินใจดังนี้
1. ชื่อ นามสกุล 
2. อายุ (ถ้าอายุอยูาระหว่าง 13-19 ปี จะตอบว่าเป็น "วัยรุ่น" แต่ถ้าไม่ใช่ ถ้าน้อยกว่าจะตอบว่าเป็น "เด็ก" และถ้ามากกว่าจะตอบว่าเป็น "วัยผู้ใหญ่"




```python
name = input("ชื่อ : ")
age = int(input("อายุ : "))
if (age <= 19 and age >= 13):
    print(name)
    print("วัยรุ่น")
elif (age > 19):
    print(name)
    print("วัยผู้ใหญ่")
elif (age < 13):
    print(name)
    print("เด็ก")
```
